# Python struct

## 用处

    1.按照指定格式将Python数据转换为字符串,该字符串为字节流,如网络传输时,不能传输int,此时先将int转化为字节流,然后再发送;
    2.按照指定格式将字节流转换为Python指定的数据类型;
    3.处理二进制数据,如果用struct来处理文件的话,需要用’wb’,’rb’以二进制(字节流)写,读的方式来处理文件;
    4.处理c语言中的结构体;

    For example:
    比如有一个报文头部在C语言中是这样定义的

    struct header
    {
          unsigned short  type;
          char[4]         tag;
          unsigned int    version;
          unsigned int    length;
    };

```python
import struct

str = struct.pack('B4sII', 0x04, b'aaaa', 0x01, 0x0e)
print(str)

types, tag, version, length = struct.unpack('B4sII', str)
print(types, tag, version, length)
```
