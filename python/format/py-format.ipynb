# Python string format

    字符串的格式化方法分为两种，分别为占位符(%)和format方式。
    占位符方式在Python2.x中用的比较广泛，随着Python3.x的使用越来越广，format方式使用的更加广泛。

## PlaceHolder

```python
from IPython.display import Image
Image(filename='../images/placeholder.png', width=800, height=600)
```

```python
name = 'openxzx'
age = 30
print("My name is %s. My age is %d" % (name, age))
```

## Format

```python
from IPython.display import Image
Image(filename='../images/format1.png', width=800, height=600)
```

```python
print("{}:{}".format('192.168.0.100',8888))
```

```python
from IPython.display import Image
Image(filename='../images/format2.png', width=800, height=600)
```

```python
print("{server}{1}:{0}".format(8888,'192.168.1.100',server='Web Server Info:'))
```

```python
from IPython.display import Image
Image(filename='../images/format3.png', width=800, height=600)
```

```python
print("{0[0]}.{1[0]}".format(('www','baidu','com'), ('sss', 'com')))
```

```python
from IPython.display import Image
Image(filename='../images/format4.png', width=800, height=600)
```

```python
print("{0}*{1}={2:0>2}".format(3,2,2*3))
print("{:*^30}".format('centered'))
print("{:*<30}".format('centered'))
```

```python
from IPython.display import Image
Image(filename='../images/format5.png', width=800, height=600)
```

```python
print("{:.3f}".format(3.1415))
print("{:.10f}".format(3.1415))
```
