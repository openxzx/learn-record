## IIC协议

    总线信号：
    SDA： 串行数据线
    SCL： 串行时钟

    总线空闲状态：
    SDA： 高电平
    SCL： 高电平

## 同步和异步

    同步是指：发送方发出数据后，等接收方发回响应以后才发下一个数据包的通讯方式。  
    异步是指：发送方发出数据后，不等接收方发回响应，接着发送下个数据包的通讯方式。
    
    SPI，IIC为同步通信，UART为异步通信。
