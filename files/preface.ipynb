# 前言

    从大学毕业到现在一直都在技术这行，也许是因为性格的问题吧，更多的偏内向，所以感觉自己根本做不了销售方面的工作。截止到现在已经工作有10年了，由于所学专业的原因，一直从事嵌入式开发这个方面包括软件和硬件。
    
    我在这里先说一下硬件吧，因为现在已经不再搞硬件设计了，内容相对较少一些吧。刚毕业时，在公司会做硬件原理图和PCB设计，使用的硬件设计软件为Altium Designer，简称AD。记得第一次做稍微复杂一点的PCB的时候，改板有5-6次吧，虽然老板没有正面责备我，但是自己也感到很是郁闷。终于，PCB尺寸上成功了，到了软件调试，使用的C语言实现的开发，跑在S3C2440开发板上。这是我人生中的第一个开发吧，真的很不容易啊。慢慢不再想搞硬件设计了，或许是真的不擅长硬件设计吧，或许是自己做事情比较马虎而硬件对细心要求更高吧，或许是自己慢慢喜欢上做软件的原因吧。

    从放弃硬件开发后，开发专门做嵌入式软件了，到现在为止，也算是会几种编程语言了吧，如熟悉C、Python、Shell、Makefile、Lua等， 了解C++、JavaScript等。再有就是在Linux系统下做软件开发，各种工具命令处理一些事情显得很方便快捷，所以还是很喜欢在Linux系统在做开发的，Linux系统下的Vim编辑器是非常强大的，同时可以使用编写Makefile来完成处理器或MCU的编译，这样会对整个的生成可执行文件的过程会比较清楚明朗。
    
    开始从事嵌入式开发的时候，都是使用C语言做开发并使用IDE集成开发环境来完成开发的编辑、编译和调试。慢慢的后来用上了Linux系统，使用Vim做编辑器，编写Makefile文件完成代码的编译，使用Git和Github来管理软件版本，感觉还是方便快捷的，所以感觉以后的方向就是这个了。
     
    随着后续开发知识面的扩展，需要编写自动化脚本、测试脚本和数据处理，慢慢的用Shell、Sed和Awk到Python2、Python3和IPython的使用，感觉到脚本语言的快捷、方便和强大，一个这样的过程，慢慢的感觉Python可以做很多事情，如数据处理（字符串显示，数据2D和3D显示等）、硬件通讯接口测试（RS232、RS485、IIC通讯等）、Web开发等等。
    
    虽然工作有很多年了，按理论来说也积累很多的工作经验和知识了，但是有时候总是感觉很多东西都是一知半解，或许是没有总结和详细考虑过的原因吧，突然有个想法想要对所学习的或者我使用的知识做个记录并上传自己的Github上，第一，可以把所学习的知识做一个总结记录下来； 第二，可以练习一下自己的写作能力，因为我的写作能力直接还是很差的，随便可以提升一下吗！ 但是之前一直都没有找到一个合适的工具，起初使用Markdown来做一些简单的记录，但是总是感觉Markdown需要记很多使用方法，说实话Markdown还是很强大的，真正原因可能是自己能力问题吧，所以我就是再次寻找其他的编写工具，突然有一天发现了Jupyter notebook工具，开始只是一知半解的看了一下这个工具，但是也没有真正的理解它。后来，有时间就看一下相关的文档和看了别人在开源Repo中使用这个工具来写文档并可以在Github中直接打开，意识到我一直找的编写文档工具就是Jupyter notebook了，在多次的尝试使用中学会了初步的使用，并且本篇文章是我使用Jupyter notebook编写的第一篇文章。
    
    我把这个Repo定义为public状态，也算是自己作为开源开发人员的一个开始吧，因为我从很多开源开发人员和开源项目中学到了很多知识，或许我的文章对大家看了并没有帮助，那就让她作为我学习的开始也挺好啊。
